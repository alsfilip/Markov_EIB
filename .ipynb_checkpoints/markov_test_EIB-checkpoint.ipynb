# Behavior of information bottleneck in different environments

We've noticed that there are some differences in behavior when the EIB is computed for a discrimination vs. a prediction task. In the discrimination task everyone seems to be on the bound, regardless of complexity, whereas in the prediction task everyone seems to fall farther from the bound as they get more complex.

This workbook is meant to explore the behavior of different models in different environments. It will start with a simple markov chain and move on from there.

First, lets load our workspace and define our general utility functions

```python
%pylab
%matplotlib inline
import pandas as pd
import numpy as np
from numba import jit
import scipy
from scipy import stats
from scipy.stats import spearmanr, pearsonr,entropy
#%config InlineBackend.figure_format = 'svg'
import glob, pickle
import seaborn as sns

import sys, csv
sys.path.append('./embo')
from embo.embo import empirical_bottleneck as eb

#Utility functions
#Load data and helper functions
def get_marginal(x):
    """
    Helper function to compute and return marginal probability distribution for a 1d vector (x)
    """
    px = np.array([np.sum(x==xi) for xi in np.sort(np.unique(x))])/len(x)
    return(px)

def get_joint(x, y):
    """
    Computes joint probability distribution between 1d vectors x and y (assumes x and y are the same length)
    """
    #  set up dictionary for joint distribution (x-->y-->freq)
    joint_x_y = {}
    
    for x_un in np.unique(x):
        joint_x_y[x_un] = dict(zip(np.unique(y), np.zeros(len(np.unique(y)))))
        
#    populate dictionary 
    for trial, x_val in enumerate(x):
        y_val = y[trial]
        joint_x_y[x_val][y_val] += 1
        
#   normalize to make distirbution 
    # Weird function that calculates the sum of all of the elements in a dictionary
    joint_sum = sum(sum(list(c.values())) for c in list(joint_x_y.values()))
    
    # Normalize - probably a better way of doing this with matrices
    for key1 in joint_x_y:
        for key2 in joint_x_y[key1]:
            joint_x_y[key1][key2] /= joint_sum
            
    return(joint_x_y)

def mutual_inf(x, y):
    """
    Calculates and returns the mutual information I(x;y)
    Assuming x,y are both [n x 1] dimensional
    """  
#     Calculate marginal distributions
    px = get_marginal(x)
    py = get_marginal(y)
    
    
    joint_x_y = get_joint(x,y)
# calculate mutual information
    mi = 0
    
    for n_x, x_un in enumerate(np.unique(x)):
        pxi = px[n_x] # p(x)
        
        for n_y, y_un in enumerate(np.unique(y)):
            pyi = py[n_y] # p(y)            
            
            joint_i = joint_x_y[x_un][y_un] # P(x,y)
            
            if ((pxi == 0) or (pyi == 0) or (joint_i ==0 )):
                continue
            else:
                mi += joint_i * np.log2(joint_i/(pxi*pyi))
                
    return mi

def deltaBound(ib_ipast,ib_ifuture,p_ipast,p_ifuture):
    ''' Function to calculate distance from the bound between an empirical IB and participant predictive info
    ib_ipast: ipast of empirical IB (x of convex hull)
    ib_ifuture: ifuture of empirical IB (y of convex hull)
    p_ipast: participant ipast (uncorrected)
    p_ifuture: participant ifuture (uncorrected)
    
    Returns participant ifuture minus the empirical bound (more negative = farther away from the bound)
    '''
    ind = np.argwhere(np.array(ib_ipast) > p_ipast)[0][0]
    slp = (ib_ifuture[ind]-ib_ifuture[ind-1])/(ib_ipast[ind]-ib_ipast[ind-1])
    intercept = ib_ifuture[ind]-(slp*ib_ipast[ind])
    #Return distance between participant Ifuture and interpolated bound - higher 
    return p_ifuture - ((p_ipast*slp)+intercept)
import numpy as np


def genMarkovObs(n,p,h):
    '''
    General purpose function to generate markov observations. Generates samples x from state z with transition probabilities h
    INPUT:
        n: number of desired observations
        p: bernoulli probability that sample from z will be the same as z
        h: transition probability of z for successive trials
    OUTPUT:
        z: sampled sources
        x: sampled observations
    '''
    z = np.zeros(n)-1
    x = np.zeros(n)-1
    zs = np.array([0,1])
    currZ = np.random.choice(zs)
    for i in np.arange(n):
        
        # Apply transition probability
        if np.random.uniform() < h:
            currZ = zs[zs!=currZ]
        z[i] = currZ
        
        # Generate sample
        if np.random.uniform() < p:
            x[i] = currZ
        else:
            x[i] = zs[zs!=currZ]
    
    return(z,x)

def getMarkovFeat(z,x):
    '''
    Function to convert observations and sources into features
    INPUT:
        z: 1 x n vector of trial by trial sources
        x: 1 x n vector of trial by trial observations
    OUTPUT:
        feat: 1 x n vector of trial by trial features
    '''
    base = np.array([2,2])**np.array([0,1])
    return(np.array([np.sum(base*np.array([z[i],x[i]])) for i in np.arange(len(z))]))

def getMarkovEstimator(x,p,h,hardmax=False):
    '''
    Function to generate responses from a model attempting to estimate state
    INPUT:
        x: 1 x n sequence of observations
        p: internal estimate of Bernoulli probability between z and x
        h: internal estimate of transition probability
    OUTPUT:
        prior_resp: model's prediction of the state z using prior prob
        post_resp: model's estimate of the state z using posterior prob (after having seen the observation)
    '''
    prior_resp = np.zeros(len(x))-1
    post_resp = np.zeros(len(x))-1
    z_est = np.array([.5,.5])
    liks = np.array([[p,1-p],[1-p,p]])
    for i,xi in enumerate(x):
        
        #Apply hazard to prior
        z0 = z_est[0].copy()
        z1 = z_est[1].copy()
        z_est[0] = z0*(1-h) + z1*h
        z_est[1] = z1*(1-h) + z0*h
        
        #Make prediction (prior to seeing observation)
        if z_est[1] == .5:
            prior_resp[i] = np.random.choice([0,1])
        else:
            if hardmax:
                prior_resp[i] = int(z_est[1]>.5)                    #Hardmax response
            else:
                prior_resp[i] = int(np.random.uniform()<z_est[1])   #Stochastic response
        
        # Update adfter seeing observation
        z_est = z_est*liks[int(xi)]
        z_est /= np.sum(z_est)
        
        #Make discrimination judgement (after seeing evidence)
        if z_est[1] == .5:
            post_resp[i] = np.random.choice([0,1])
        else:
            if hardmax:
                post_resp[i] = int(z_est[1]>.5)                    #Hardmax response
            else:
                post_resp[i] = int(np.random.uniform()<z_est[1])   #Stochastic response       
        
    return(prior_resp,post_resp)

def getPatternEstimator(x):
    '''
    Function to generate responses from model that uses info from 2 trials back to make its responses
    '''
    prior_resp = np.zeros(len(x))-1
    for i in np.arange(len(x)):
        if i < 2:
            r = np.random.choice([0,1])
        else:
            r = int(x[i-2] == x[i-1])
        prior_resp[i] = r
    
    post_resp = prior_resp.copy() #figure this out later
    return(prior_resp,post_resp)

def genSims(n,p,discr=False):
    mkov_sims = {}
    h_ex = [0.05,0.1,0.2,0.3,0.5,0.7,0.8,0.9,0.95]
    for i,h in enumerate(h_ex):

        # Dictionaries to save input
        mkov_sims[h] = {}

        # Generate observations
        z,x = genMarkovObs(n,p,h)

        # Convert to features
        if discr:
            feat = getMarkovFeat(z[:-1],x[1:])
        else:
            feat = getMarkovFeat(z,x)
        feat_patt = getMarkovFeat(x[:-1],x[1:])
        

        # Get empirical bottleneck values
        if discr:
            eib_ip,eib_if,eib_beta,eib_mi,eib_hx,eib_hy = eb(feat,z[1:],numbeta=1000,maxbeta=100,parallel=8)
        else:
            eib_ip,eib_if,eib_beta,eib_mi,eib_hx,eib_hy = eb(feat[:-1],z[1:],numbeta=1000,maxbeta=100,parallel=8)
        mkov_sims[h]['eib_ipast'] = eib_ip
        mkov_sims[h]['eib_ifuture'] = eib_if
        mkov_sims[h]['bound'] = eib_mi

        # Get responses
        #Ideal observer
        prior_ideal, post_ideal = getMarkovEstimator(x,p,h)
        
        #Ideal observer - hardmax
        prior_ideal_hm, post_ideal_hm = getMarkovEstimator(x,p,h,hardmax=True)
        
        #Simple model (repeat previous state)
        prior_simple = np.concatenate((np.array([np.random.choice([0,1])]),x[:-1]),axis=None)
        post_simple = x.copy()

        #Overestimator
        prior_over, post_over = getMarkovEstimator(x,p,.99999)

        #Underestimator
        prior_under, post_under = getMarkovEstimator(x,p,.00001)
        
        #Pattern estimator
        prior_patt, post_patt = getPatternEstimator(x)

        
        #Decide whether or not we want predictive or discrimination task
        if discr:
            resps = [post_ideal,post_ideal_hm,post_simple,post_over,post_under,post_patt]
        else:
            resps = [prior_ideal,prior_ideal_hm,prior_simple,prior_over,prior_under,prior_patt]

        #Save simulation predictive information
        models = ['ideal','ideal_hardmax','simple','over','under','pattern']
        for j,mod in enumerate(models):
            r = resps[j]
            if mod == 'pattern':
                mkov_sims[h][mod+'_ipast'] = mutual_inf(feat_patt[:-1],r[2:])
            else:
                if discr:
                    mkov_sims[h][mod+'_ipast'] = mutual_inf(feat,r[1:])
                else:
                    mkov_sims[h][mod+'_ipast'] = mutual_inf(feat[:-1],r[1:])
            mkov_sims[h][mod+'_ifuture'] = mutual_inf(r,z)
          
    return(mkov_sims)

def pltSims(mkov_sims):
    sp = 1
    for key in mkov_sims.keys():
        subplot(2,5,sp)
        plot([-.05,1.05], [0,0],'--k')
        plot([mkov_sims[key]['bound'],mkov_sims[key]['bound']],[0,mkov_sims[key]['bound']],'--k',alpha=.5)
        plot([0,mkov_sims[key]['bound']],[mkov_sims[key]['bound'],mkov_sims[key]['bound']],'--k',alpha=.5)
        plot(mkov_sims[key]['eib_ipast'],mkov_sims[key]['eib_ifuture'],'-k')
        plot(mkov_sims[key]['ideal_ipast'],mkov_sims[key]['ideal_ifuture'],'o',label='Ideal',markersize=9)
        #plot(mkov_sims[key]['ideal_hardmax_ipast'],mkov_sims[key]['ideal_ifuture'],'o',label='Ideal - Hardmax',markersize=9)
        plot(mkov_sims[key]['simple_ipast'],mkov_sims[key]['simple_ifuture'],'o',label='Simple')
        plot(mkov_sims[key]['over_ipast'],mkov_sims[key]['over_ifuture'],'o',label='Overestimator')
        plot(mkov_sims[key]['under_ipast'],mkov_sims[key]['under_ifuture'],'o',label='Underestimator')
        plot(mkov_sims[key]['pattern_ipast'],mkov_sims[key]['pattern_ifuture'],'o',label='Pattern')
        title('$h$ = %s'%str(key),size=18)
        ylim([-.03,1])
        xlim([-.03,1.03])
        if(sp in [1,6]):
            yticks([0,.5,1])
            ylabel('$I_{future}$',size=15)
        else:
            yticks([])

        if(sp < 5):
            xticks([])
        else:
            xlabel('$I_{past}$',size=15)
        if sp == 5:
            legend()
        sp+=1
```


## 1) Empirical bottleneck for a markov chain

We begin by generating a markov chain that can take on 2 states $z\in\{z_1,z_2\}$, and transitions between states on the next trial with probability $h$:

$$z_t=\left\{
\begin{array}{c l}	
     1-h,z_t = z_{t-1}\\
     h, z_t \neq z_{t-1}
\end{array}\right.$$

The predictive information for a simple markov chain is easy to compute. Since the last observation contains all of the past, we can compute the predictive informaiton of a markov chain as the mutual information between observations at time $t-1$ and $t$

$$ I(z_{t-1};z_{t}) = \sum_{z_{t-1}}\sum_{z_t}p(z_{t-1},z_t)\log_2 \Big[\frac{p(z_{t-1},z_t)}{p(z_{t-1})p(z_t)}\Big]$$

If we assume that the transitions between both states are equal and symmetric, the entries of the joint distribution $p(z_{t-1},z_t)$ are equal to $h$ when $z_t \neq z_{t-1}$ and $(1-h)$ when $z_t = z_{t-1}$. Predictive information concerns the information the past carries about the future. In a markov chain, all of the past is contained within the current state. The predictability of the next state is determined by the transition probability $h$, which can be treated as Bernoulli random variable. As a result, the predictive information for any markov chain with two states and equal and symmetric transition probabilities is equavalent to $1-S(h)$, where $S$ denotes the Shannon entropy.

$$ S(h) = -\sum_h h\log_2h$$
$$ I(z_t;z_{t-1}) = 1-S(h)$$

```python
# Plot relationship between shannon entropy and predictive information
hs = np.linspace(.01,1,100)
h_ex = np.array([0.05,.1,.2,.3,.5,.7,.8,.9,.95])
sh = np.array([-(h*np.log2(h)) - ((1-h)*np.log2(1-h)) for h in hs])
sh_ex = np.array([-(h*np.log2(h)) - ((1-h)*np.log2(1-h)) for h in h_ex])

figure(1,(15,4))
subplot(1,2,1)
plot(hs,sh)
plot(h_ex,sh_ex,'o')
ylabel('$S(h)$',size=18)
xlabel('Transition Probability ($h$)',size=18)

subplot(1,2,2)
plot(hs,1-sh)
plot(h_ex,1-sh_ex,'o')
ylabel('Predictive Information\n$I(z_t;z_{t-1})$',size=18)
xlabel('Transition Probability ($h$)',size=18)
ylim([-.03,1])
```

Now lets see how different agents compare in the information bottleneck space. We'll start with two simple agents:
- 1) Ideal observer - knows the transition probability and does bayesian inference accordingly
- 2) Simple observer - does not attempt to do any inference and instead just uses the last observed state as its prediction for the next state.

For the simulations below we'll plot the information bottleneck bound for the different values of $h$ highlighted in orange above. The saturation point for each curve should correspond to the predictive information values depicted in the right panel. The curve shows the relationship between encoding and predictive accuracy, where it show the extent to which encoded information can be used to acurately predict the future.

We will plot the information bottleneck values for each simulation (ideal and simple observers) to see how close each simulation falls from the curve. This is computed by measuring the extent to which past observations are being encoded by each simulation ($I_{past}$), and comparing them to how close in predictive accuracy ($I_{future}$) each simulation falls on the bound. Given the markovian nature of the models and stimuli, the amount of information being encoded is measured as

$$I_{past} = I(z_{t-1};r_t) = \sum_z\sum_r p(z_{t-1},r_t)\log_2\Big[\frac{p(z_{t-1},r_t)}{p(z_{t-1})p(r_t)}\Big]$$

Since the models are attempting to predict the state $z_t$, the predictive accuracy, or effectiveness of each model is measured as

$$I_{future}=I(z_t;r_t) = \sum_z\sum_r p(z_t,r_t)\log_2\Big[\frac{p(z_t,r_t)}{p(z_t)p(r_t)}\Big]$$

The closer this predictive accuracy falls to the bound for a given level of $I_{past}$, the more optimal the model.

There are 5 models bering cosidered below:

- 1) Ideal observer: Bayesian model that does inference according to the correct transition probability. Note that the responses are probabilistic (i.e., responses are made as a function of the prior probability that the model thinks the next state will be 0 or 1).

- 2) Simple: Model that repeats the last observed state as it's prediction.

- 3) Overestimator: Bayesian model which assumes a really high transition probability (.9999)

- 4) Underestimator: Bayesian model which assumes a really low transition probability (.0001)

- 5) Pattern: Model that responds 1 when the observed stimuli on the last 2 trials are the same, and 0 when they differ. Used as an example of a model that encodes lots from the task, but does a poor job predicting the next state.

```python
figure(2,(18,10))
mkov_sims = genSims(10000,1)
pltSims(mkov_sims)
```

There are a couple of interesting things to note:

- 1) All of the models that have the correct idea about the markovian nature of the stimuli (all except the simple and pattern models) fall on the bound. 

- 2) While the optimal model is on the bound, its predictive accuracy is lower than either the over/underestimator models, or even the simple model.  

- 3) Not too sure why both the under and overestimator models have such high predictive accuracy... 

## 2) Hidden markov model - prediction

Now we're going to run the same analyses but with stimuli generated from a hidden markov model. Here the transition probabilities are the same, but the observations $x$ are noisy representations of the current state $z$. In this particular example with probability $p$:

$$x_t=\left\{
\begin{array}{c l}	
     p,z_t\\
     1-p, x_t \neq z_t
\end{array}\right.$$

In this instance, doing proper inference really matters matters the observations are a noisy representation of the current state.

```python
figure(2,(18,10))
hmkov_sims = genSims(10000,.8)
pltSims(hmkov_sims)
```

Interesting differences between this and the Markov case:

- 1) Now that there is a latent state to be inferred, the simple model does very poorly (encodes a lot of information, i.e., the observed variable, but uses this ineffectively). The ideal model is much closer to the bound.

- 2) That said, I find it interesting that the ideal observer is close, but still slightly off the bound. Additionally, note how the under and over estimators change as a function of the transition probability. The understimator seems like a more optimal model when H < .5, and the over estimator more optimal when H > .5
